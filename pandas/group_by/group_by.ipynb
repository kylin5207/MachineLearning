# group_by按某列分组

例如：按照A字段进行分组

```python
import pandas as pd
```

### 1. 生成数据

```python
data = {
    'A': ['a', 'c', 'b', 'c', 'a'],
    'B': [0, 1, 3, 2, 3],
    'C': [1.2, 3.2, 4.2, 2.1, 1.0]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

### 2. 按A进行分组

#### (1)首先通过groupby得到DataFrameGroupBy对象, 比如data.groupby('A')

#### (2) 然后选择需要研究的列， 返回的是一个SeriesGroupBy对象，不易直观查看

```python
df.groupby("A")["B"]
```

```python
pd.DataFrame(df.groupby("A")["B"])
```

#### (3) 对SeriesGroupby进行操作, 比如.mean(), 相当于对每个组的Series求均值

```python
df.groupby("A")["B"].agg('max')
```

```python
df.groupby("A")["B"].agg('count')
```

```python
df.groupby("A")["B"].agg('sum')
```

```python
df.groupby("A")["B"].agg('size')
```
